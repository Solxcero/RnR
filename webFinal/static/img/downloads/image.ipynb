````python
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time
import urllib.request
import os
import pandas as pd
````

````python
def download_images(search_term, download_dir):
    if not os.path.exists(download_dir):
        os.makedirs(download_dir)

    driver = webdriver.Chrome('./chromedriver')
    driver.get("https://www.google.com/imghp?hl=en")

    search_bar = driver.find_element(By.NAME,"q")
    search_bar.send_keys(search_term)
    search_bar.send_keys(Keys.RETURN)

    driver.implicitly_wait(5)
    images = driver.find_elements(By.CSS_SELECTOR,".rg_i.Q4LuWd")
    count = 1
    for image in images:
        if count > 20:
            break
        try:
            image.click()
            driver.implicitly_wait(5)
            time.sleep(5)
            imgUrl = driver.find_element(By.XPATH,'//*[@id="Sva75c"]/div[2]/div/div[2]/div[2]/div[2]/c-wiz/div/div[1]/div[2]/div[2]/div/a/img')
            src = imgUrl.get_attribute('src')
            urllib.request.urlretrieve(src, os.path.join(download_dir, search_term + "_" + str(count) + ".jpg"))
            print("Image saved: {}_{}.jpg".format(search_term, count))
            count += 1
        except:
            pass

    driver.close()
````

````python
def download_images_for_hotels(hotel_df, download_dir):
    for i, row in hotel_df.iterrows():
        search_term = row["ht_name"]
        download_images(search_term, os.path.join(download_dir, str(row["ht_id"])))
````
