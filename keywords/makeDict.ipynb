```python
for idx, cat in enumerate(rev_cat):
    df = rev_df2.copy()
    df['cat_id'] = idx+1
    df[cat] = 0
    for i in range(rev_df2.shape[0]):
        for c in rev_cat[cat] :
            if c in df.iloc[i,2] :
                df.iloc[i,4] = 1
                break    
    df = df[df[cat]==1]   
    df = df[['ht_id','cat_id','date','review']]
    df.to_csv(f'./files/{cat}.csv',index=False) 
```
